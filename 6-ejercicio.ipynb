```python
import matplotlib.pyplot as plt

# Datos proporcionados
rubros = ['Alimentos, bebidas y Tabaco', 'Vestido y calzado', 'Vivienda y combustibles',
          'Artículos y servicios para la canasta', 'Transporte y comunicaciones', 
          'Educación y esparcimiento', 'Cuidados de la salud', 'Transferencia de gasto', 
          'Autoconsumo', 'Remuneraciones en especie', 'Transferencias en especie', 
          'Estimación del alquiler de la vivienda']

gastos = [7389, 1155, 2207, 1316, 4052, 4052, 1530, 676, 305, 545, 2352, 4091]

# Ordenamos los rubros y gastos de mayor a menor
rubros_gastos = list(zip(rubros, gastos))
rubros_gastos.sort(key=lambda x: x[1], reverse=True)
rubros, gastos = zip(*rubros_gastos)

# Calculamos los porcentajes
total_gastos = sum(gastos)
porcentajes = [(gasto / total_gastos) * 100 for gasto in gastos]

# Creamos el gráfico de pastel
fig, ax = plt.subplots(figsize=(10, 8))
wedges, labels, autopct = ax.pie(porcentajes, labels=rubros, autopct='%1.1f%%', pctdistance=0.9, startangle=140)
plt.setp(wedges, edgecolor='white')

# Cambiamos el color anaranjado por dorado


# Cambiamos el color azul por gris
wedges[11].set_facecolor('pink')

wedges[10].set_facecolor('gold')

# Ajustamos el aspecto del gráfico
ax.axis('equal')
plt.title('Distribución de gastos por rubro')
plt.show()
```
